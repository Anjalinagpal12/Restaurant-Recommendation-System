```python
df_new = df_new.dropna()
```

```python
df_new = pd.get_dummies(df_new, columns=['Category', 'Locality'])

similarity_matrix = cosine_similarity(df_new.drop(columns=['Restaurant_Name', 'Dining_Rating']))

similarity_df = pd.DataFrame(similarity_matrix, index=df_new['Restaurant_Name'], columns=df_new['Restaurant_Name'])
```

```python
def recommend_restaurants(Restaurant_Name, num_recommendations=5):
    if Restaurant_Name not in similarity_df.index:
        return "Restaurant not found in the dataset."
    
    similar_restaurants = similarity_df[Restaurant_Name].sort_values(ascending=False)
    
    
    recommendations = similar_restaurants.head(num_recommendations + 1).index.tolist()
    
    
    recommendations.remove(Restaurant_Name)
    
    return recommendations

# Test halwai
print(recommend_restaurants('Cafe Wink'))
```

```python
price_features = df_new[['Pricing_for_2']]
similarity_matrix = cosine_similarity(price_features)

similarity_df = pd.DataFrame(similarity_matrix, index=df_new['Restaurant_Name'], columns=df_new['Restaurant_Name'])
```

```python
def recommend_restaurants_by_price_range(min_price, max_price, num_recommendations=5):
    
    min_price = float(min_price)
    max_price = float(max_price)
    
    filtered_data = df_new[(df_new['Pricing_for_2'] >= min_price) & (df_new['Pricing_for_2'] <= max_price)]
    
    
    if len(filtered_data) > num_recommendations:
        recommendations = filtered_data.sample(num_recommendations)
    else:
        recommendations = filtered_data
    
    return recommendations['Restaurant_Name'].tolist()

# PAISA HALWAI
print(recommend_restaurants_by_price_range(500,2000))
```

```python
def recommend_restaurants_by_rating(min_rating, num_recommendations=5):
    
    min_rating = float(min_rating)
    
    
    filtered_data = df_new[df_new['Dining_Rating'] >= min_rating]
    
    
    if len(filtered_data) > num_recommendations:
        recommendations = filtered_data.sample(num_recommendations)
    else:
        recommendations = filtered_data
    
    return recommendations['Restaurant_Name'].tolist()


min_rating = 4.5
recommendations = recommend_restaurants_by_rating(min_rating)
print(f"Recommendations for restaurants with a rating of {min_rating} or higher:")
print(recommendations)
```
